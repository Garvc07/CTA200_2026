## Question 2

```python
import numpy as np
from scipy.integrate import solve_ivp
```

### Part 1

```python
def lorenz_equations(t, W, sigma, b, r):
    """
    Computes the Lorenz equations 25, 26, 27 outlined in assignment 3
    
    Parameters:
    t -- float representing the current time 
    W -- array that represents the current state vector [X, Y, Z]
    sigma -- float, Prandtl number (ratio of kinematic viscosity to 
    thermal diffusivity)
    b -- float, dimensionless length scale
    r -- float, Rayleigh number
    
    Returns:
    list of derivatives [dX/dt, dY/dt, dZ/dt]
    """
    X, Y, Z = W
    
    # Equations outlined in assignment pdf
    dX_dt = -sigma * (X - Y)
    dY_dt = (r * X) - Y - (X * Z)
    dZ_dt = (-b * Z) + (X * Y)
    
    return [dX_dt, dY_dt, dZ_dt]
```

```python
def lorenz_equations(t, W, sigma, r, b):
    """Computes the derivatives of the Lorenz system."""
    X, Y, Z = W
    return [-sigma * (X - Y), r * X - Y - X * Z, -b * Z + X * Y]

sigma, r, b = 10.0, 28.0, 8.0/3.0
W0 = np.array([0.0, 1.0, 0.0])

t_span = (0, 60)
delta_t = 0.01
t_eval = np.arange(t_span[0], t_span[1] + delta_t, delta_t)

sol = solve_ivp(lorenz_equations, t_span, W0, args=(sigma, r, b), 
                t_eval=t_eval, rtol=1e-10, atol=1e-11)

X, Y, Z = sol.y


# --- PART 3: Reproduce Figure 1 (Stacked Panels) ---
# Lorenz split the first 3000 iterations into three stacked graphs
fig1, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8), sharey=True)

# Iterations 0 to 1000
ax1.plot(np.arange(0, 1000), Y[0:1000], color='black', lw=1)
ax1.set_xlim(0, 1000)

# Iterations 1000 to 2000
ax2.plot(np.arange(1000, 2000), Y[1000:2000], color='black', lw=1)
ax2.set_xlim(1000, 2000)

# Iterations 2000 to 3000
ax3.plot(np.arange(2000, 3000), Y[2000:3000], color='black', lw=1)
ax3.set_xlim(2000, 3000)

fig1.suptitle("Figure 1: Time Evolution of Y (Iterations 0 - 3000)")
ax3.set_xlabel("Iteration Number (N = t / Δt)")
plt.tight_layout()
plt.show()
```

```python
# --- PART 4: Reproduce Figure 2 (Phase Space Projections) ---
# Lorenz specifically plotted iterations 1400 to 1900 (t=14 to t=19)
# Top panel is Z vs Y. Bottom panel is X vs Y.
start, end = 1400, 1900

fig2, (ax_top, ax_bottom) = plt.subplots(2, 1, figsize=(7, 10))

# Top: Z vertical, Y horizontal
ax_top.plot(Y[start:end], Z[start:end], color='black', lw=1)
ax_top.set_ylabel("Z")
ax_top.set_xlabel("Y")
ax_top.set_title("Projections on Y-Z plane")

# Bottom: X vertical, Y horizontal
ax_bottom.plot(Y[start:end], X[start:end], color='black', lw=1)
ax_bottom.set_ylabel("X")
ax_bottom.set_xlabel("Y")
ax_bottom.set_title("Projections on X-Y plane")

plt.tight_layout()
plt.show()
```

```python
# --- PART 5: Sensitive Dependence (The Butterfly Effect) ---
W0_prime = W0 + np.array([0.0, 1e-8, 0.0])

# Solve again with exact same strict tolerances
sol_prime = solve_ivp(lorenz_equations, t_span, W0_prime, args=(sigma, r, b), 
                      t_eval=t_eval, rtol=1e-10, atol=1e-11)

X_diff = sol.y[0] - sol_prime.y[0]
Y_diff = sol.y[1] - sol_prime.y[1]
Z_diff = sol.y[2] - sol_prime.y[2]
distance = np.sqrt(X_diff**2 + Y_diff**2 + Z_diff**2)

plt.figure(figsize=(10, 5))
plt.semilogy(sol.t, distance, color='red', lw=1)
plt.title("Exponential Divergence of Nearby Trajectories")
plt.xlabel("Time (t)")
plt.ylabel("Distance (Log Scale)")
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.xlim(0, 50) # Zooming in to see the linear growth clearly
plt.show()
```

### Part 2

```python
sigma = 10.0
r = 28.0
b = 8.0/3.0
W0 = np.array([0.0, 1.0, 0.0]) 
t_span = (0, 60) # Integrate from t=0 to t=60
del_t = 0.01

t_eval = np.arange(t_span[0], t_span[1] + del_t, del_t)

#solution = solve_ivp(lorenz_equations, t_span, W0, args=(sigma, r, b), 
                #t_eval=t_eval, dense_output=True)

solution = solve_ivp(lorenz_equations, t_span, W0, args=(sigma, r, b), 
                t_eval=t_eval, rtol=1e-10, atol=1e-11)
```

### Part 3

```python
N = solution.t / del_t
X, Y, Z = solution.y

plt.figure(figsize=(10, 4))
plt.plot(N, Z, color='blue', linewidth=0.8)
#plt.title("Lorenz Figure 1 Reproduction: Time Evolution of Y")
#plt.xlabel("Iteration Number (N = t / Δt)")
#plt.ylabel("Y")
#plt.grid(True)
#plt.show()
```

```python
# --- PART 3: Reproduce Figure 1 (Stacked Panels) ---
# Lorenz split the first 3000 iterations into three stacked graphs
fig1, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8), sharey=True)

# Iterations 0 to 1000
ax1.plot(N[0:1000], Y[0:1000], color='black', lw=1)
ax1.set_xlim(0, 1000)

# Iterations 1000 to 2000
ax2.plot(np.arange(1000, 2000), Y[1000:2000], color='black', lw=1)
ax2.set_xlim(1000, 2000)

# Iterations 2000 to 3000
ax3.plot(np.arange(2000, 3000), Y[2000:3000], color='black', lw=1)
ax3.set_xlim(2000, 3000)

fig1.suptitle("Figure 1: Time Evolution of Y (Iterations 0 - 3000)")
ax3.set_xlabel("Iteration Number (N = t / Δt)")
plt.tight_layout()
plt.show()
```

### Part 4

```python
# --- PART 4: Figure 2 (Phase Portrait for t=14 to 19) ---
# Create closely spaced time intervals
t_fine = np.linspace(14, 19, 1000)

# Evaluate our continuous solution at these fine intervals
W_fine = solution.sol(t_fine)
X_fine = W_fine[0]
Y_fine = W_fine[1]

plt.figure(figsize=(6, 6))
plt.plot(X_fine, Y_fine, color='purple')
plt.title("Lorenz Figure 2 Reproduction: Phase Space (Y vs X)")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
# --- PART 4: Reproduce Figure 2 (Phase Space Projections) ---
# Lorenz specifically plotted iterations 1400 to 1900 (t=14 to t=19)
# Top panel is Z vs Y. Bottom panel is X vs Y.
start, end = 1400, 1900

fig2, (ax_top, ax_bottom) = plt.subplots(2, 1, figsize=(7, 10))

# Top: Z vertical, Y horizontal
ax_top.plot(Y[start:end], Z[start:end], color='black', lw=1)
ax_top.set_ylabel("Z")
ax_top.set_xlabel("Y")
ax_top.set_title("Projections on Y-Z plane")

# Bottom: X vertical, Y horizontal
ax_bottom.plot(Y[start:end], X[start:end], color='black', lw=1)
ax_bottom.set_ylabel("X")
ax_bottom.set_xlabel("Y")
ax_bottom.set_title("Projections on X-Y plane")

plt.tight_layout()
plt.show()
```

### Part 5

```python
# --- PART 5: Sensitive Dependence on Initial Conditions ---
# Perturb the initial condition using element-by-element addition
W0_prime = W0 + np.array([0.0, 1e-8, 0.0])

# Solve again with the perturbed condition
sol_prime = solve_ivp(lorenz_equations, t_span, W0_prime, args=(sigma, r, b), t_eval=t_eval)

# Calculate the Euclidean distance between W and W' at every time step
# Distance = sqrt( (X - X')^2 + (Y - Y')^2 + (Z - Z')^2 )
X_diff = solution.y[0] - sol_prime.y[0]
Y_diff = solution.y[1] - sol_prime.y[1]
Z_diff = solution.y[2] - sol_prime.y[2]
distance = np.sqrt(X_diff**2 + Y_diff**2 + Z_diff**2)

# Semilog plot (linear t on x-axis, logarithmic distance on y-axis)
plt.figure(figsize=(10, 5))
plt.semilogy(solution.t, distance, color='red', linewidth = 0.2)
plt.title("Exponential Divergence of Nearby Trajectories")
plt.xlabel("Time (t)")
plt.ylabel("Distance between W and W' (Log Scale)")
plt.grid(True) # Grids for log scales look better dashed
plt.show()
```

```python
# --- PART 5: Sensitive Dependence (The Butterfly Effect) ---
W0_prime = W0 + np.array([0.0, 1e-8, 0.0])

# Solve again with exact same strict tolerances
sol_prime = solve_ivp(lorenz_equations, t_span, W0_prime, args=(sigma, r, b), 
                      t_eval=t_eval, rtol=1e-10, atol=1e-11)

X_diff = solution.y[0] - sol_prime.y[0]
Y_diff = solution.y[1] - sol_prime.y[1]
Z_diff = solution.y[2] - sol_prime.y[2]
distance = np.sqrt(X_diff**2 + Y_diff**2 + Z_diff**2)

plt.figure(figsize=(10, 5))
plt.semilogy(solution.t, distance, color='red', lw=1)
plt.title("Exponential Divergence of Nearby Trajectories")
plt.xlabel("Time (t)")
plt.ylabel("Distance (Log Scale)")
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.xlim(0, 50) # Zooming in to see the linear growth clearly
plt.show()
```
